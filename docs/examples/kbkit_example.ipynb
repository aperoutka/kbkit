# Kirkwood-Buff Analysis for Binary System

```python
from kbkit import KBPipeline, Plotter
```

## Create and run KBPipeline

```python
pipe = KBPipeline(
    # path to directory containing systems
    base_path='./test_data/systems',
    # path to pure component directory
    pure_component_path='./test_data/pure_components',
    # name for rdf file directory (available in each system)
    rdf_dir="rdf_files_npt",
    # ensemble type: (npt, nvt)
    ensemble="npt",
    # how to integrate activity coefficient derivatives: (numerical, polynomial)
    gamma_integration_type="numerical"
)

pipe.run()
```

## Create Plotter object

```python
# provide molecules mapped to desired name
molecule_map = {
    "DMDBP": "DBP",
    "NDODE": "Dodecane",
}

# which molecule to use for labels on x-axis?
x_mol = "DMDBP"


plotter = Plotter(kb_obj=pipe.kb, x_mol=x_mol, molecule_map=molecule_map)
```

#### Display available properties to plot

```python
plotter.available_properties()
```

#### Plot KBIs as a function of composition

```python
# figure for KBIs
plotter.plot_property("kbi")
```

#### Plot ln activity coefficients

```python
plotter.plot_property('lngamma')
```

#### Make all figures (stored in ./test_data/system/kb_analysis)

```python
plotter.make_figures(energy_units="kJ/mol") 
```
